```python
import pandas as pd
import numpy as np
```

```python
# box office data collected from: https://www.boxofficemojo.com/year/world/2017/?grossesOption=totalGrosses
data = pd.read_csv("BoxOffice.csv")
```

```python
data.head(10)
```

```python
data.shape
```

```python
data['Year'] = data['Year'].astype(int)
```

```python
col_names = ['Worldwide', 'Domestic', 'Foreign']
col_names2 = ['%Domestic', "%Foreign"]

for col in col_names:
    data[col] = data[col].apply(lambda x: x[1:].replace(',', ''))
    
for col in col_names2:
    data[col] = data[col].apply(lambda x: x[:-1])
```

```python
data.head()
```

```python
data.to_csv("BoxOffice_clean.csv")
```
