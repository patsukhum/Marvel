```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../raw/genre_and_reviews_omdb.csv')
```

```python
genres = df['Genre'].values
```

```python
# genres_all = [g.split(",") for g in genres ]
genres_all = []
null_idx = []
for i,g in enumerate(genres):
    if type(g) != float:
        g_arr = g.split(",")
        g_arr2 = [g.strip() for g in g_arr]
        genres_all.append(g_arr2)
    else:
#         genres_all.append(g)
        null_idx.append(i)
```

```python
sub_df = df[~df['Genre'].isna()]
```

```python
all_genres = set()
for gs in genres_all:
    for g in gs:
        all_genres.add(g)
```

```python
all_genres_dict = {}
for i,g in enumerate(all_genres):
    all_genres_dict[g] = i
```

```python
priority_genres = ['Animation','Musical','Game-Show','Reality-TV','Biography','Western','Short','Documentary','Sport', 'Film-Noir','War','History']
```

```python
priority_genres_set = set(priority_genres)
```

```python
genre_types = {
    'Action': 'Action/Adventure',
    'Adventure': 'Action/Adventure',
    'Thriller' : 'Thriller/Mystery/Horror/Crime',
    'Mystery' : 'Thriller/Mystery/Horror/Crime',
    'Horror' : 'Thriller/Mystery/Horror/Crime',    
    'Crime' : 'Thriller/Mystery/Horror/Crime',        
    'Animation' : 'Animation/Fantasy',
    'Fantasy' : 'Animation/Fantasy',    
    'Biography' : 'Biography/History/Documentary/War',
    'History' : 'Biography/History/Documentary/War',
    'Documentary' : 'Biography/History/Documentary/War',
    'War' : 'Biography/History/Documentary/War',    
    'Drama' : 'Drama/Family/Romance/Comedy', 
    'Family' : 'Drama/Family/Romance/Comedy', 
    'Romance' : 'Drama/Family/Romance/Comedy',         
    'Comedy' : 'Drama/Family/Romance/Comedy',   
    'Fantasy' : 'Fantasy/Sci-Fi',
    'Sci-Fi' : 'Fantasy/Sci-Fi'    
}
#     Others:
# 'Film-Noir',
#  'Game-Show',
#  'Music',
#  'Musical',
#  'Reality-TV',
#  'Short',
#  'Sport',
#  'Western'
```

```python
genres_all2 = []
for gs in genres_all:
    cur_g = []
    has_priority = False
    for g in gs:
        if g in priority_genres_set:
            has_priority = True
            cur_g.append(g)
    if has_priority:
        genres_all2.append(np.random.choice(cur_g))
    else:
        genres_all2.append(np.random.choice(gs))
```

```python
genres_all2
```

```python
genres_all3 = []
for v in genres_all2:
    if v not in genre_types:
        genres_all3.append('Others')
    else:
        genres_all3.append(genre_types[v])
```
