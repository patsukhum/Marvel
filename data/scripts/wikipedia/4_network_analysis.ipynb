```python
import json

import networkx as nx

from marvel import config
```

```python
edges = json.load((config.cleandata / 'all_character_links.json').open())
nodes = json.load((config.cleandata / 'all_character_nodes.json').open())
```

```python
edges
```

```python
nodes
```

```python
g = nx.DiGraph()
g.add_nodes_from([n['id'] for n in nodes])
g.add_weighted_edges_from((d['source'], d['target'], d['count']) for d in edges)

centrality = nx.eigenvector_centrality_numpy(g, weight='weight')

for i, n in enumerate(nodes):
    n.update({'centrality': centrality[i]})
```

```python
json.dump(nodes, (config.cleandata / 'all_character_nodes_centrality.json').open('w'))
```
